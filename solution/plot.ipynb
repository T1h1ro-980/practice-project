```python

import matplotlib.pyplot as plt
import torch
from train import train_model

```

```python
def plot_training_history(history, i=None):
    """Визуализирует историю обучения"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

    ax1.plot(epochs[0:i],history["avg_train_loss"][0:i],'r-', label='Train Loss')
    ax1.set_title('Loss')
    ax1.set_xlabel('Epochs')  # Название оси X
    ax1.set_ylabel('Loss')  
    ax1.legend()
    ax2.plot(epochs[0:i],history["avg_train_acc"][0:i], label='Train accuracy',)
    ax2.set_title('accuracy')
    ax2.set_xlabel('Epochs')  # Название оси X
    ax2.set_ylabel('acc')  
    ax2.legend()
    plt.tight_layout()
    plt.show()  
model = train_model()
history = model[1]
epochs = range(1, len(history["avg_train_loss"]) + 1)
for i in range(2,7):
    plot_training_history(history, i)

```
